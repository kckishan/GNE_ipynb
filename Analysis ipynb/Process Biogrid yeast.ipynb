```python
import scipy.io as sio
from scipy.sparse import coo_matrix
def _load_network(filename, mtrx='adj'):
    print ("### Loading [%s]..." % (filename))
    if mtrx == 'adj':
        i, j, val = np.loadtxt(filename).T
        A = coo_matrix((val, (i, j)))
        A = A.todense()
        A = np.squeeze(np.asarray(A))
#         if A.min() < 0:
#             print ("### Negative entries in the matrix are not allowed!")
#             A[A < 0] = 0
#             print ("### Matrix converted to nonnegative matrix.")
#             print
#         if (A.T == A).all():
#             pass
#         else:
#             print ("### Matrix not symmetric!")
#             A = A + A.T
#             print ("### Matrix converted to symmetric.")
    else:
        print ("### Wrong mtrx type. Possible: {'adj', 'inc'}")
    A = A - np.diag(np.diag(A))
#     A = A + np.diag(A.sum(axis=1) == 0)

    return A

```

```python
import pickle
from evaluation import *
from numba import jit

@jit
def read_test_link(testlinkfile):
    X_test = []
    f = open(testlinkfile)
    line = f.readline()
    while line:
        line = line.strip().split(" ")
        X_test.append([int(line[0]), int(line[1]), int(line[2])])
        line = f.readline()
    f.close()
#     print("test link number:", len(X_test))
    return X_test


```
