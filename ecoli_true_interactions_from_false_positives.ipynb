```python
import pandas as pd
import numpy as np
```

```python
biogrid_df = pd.read_csv('data/ecoli/edgelist_biogrid_name.csv', sep=",", header=None)
```

```python
biogrid_df.columns
```

```python
fp_df = pd.read_csv('data/ecoli/ecoli_prediction_false_positives.txt', sep="\t", header=0)
```

```python
fp_df.columns
```

```python
merged_1 = pd.merge(biogrid_df, fp_df, left_on=[0,1],right_on=["gene_x", "gene_y"])
```

```python
merged_2 = pd.merge(biogrid_df, fp_df, left_on=[0,1],right_on=["gene_y", "gene_x"])
```

```python
merged = pd.concat([merged_1, merged_2])
```

```python
merged = merged.iloc[:,[0,1,4]]
```

```python
merged.drop_duplicates()
```

```python
merged.to_csv("./data/ecoli/ecoli_true_interactions_from_false_positives.txt", header=None, index=None, sep=",")
```
